```python
# Cell 1: Imports & Data Loading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure plot output directory exists
os.makedirs('outputs/plots', exist_ok=True)

# Load the data
df = pd.read_csv('data/heart.csv')
```

```python
# Cell 2: Summary Statistics
display(df.describe())
```

```python
# Cell 3: Histograms for Key Features
features = ['age', 'chol', 'trestbps', 'thalach']

for feature in features:
    plt.figure(figsize=(7,4))
    sns.histplot(df[feature], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {feature}')
    plt.xlabel(feature)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.savefig(f'outputs/plots/hist_{feature}.png')
    plt.show()
```

```python
# Cell 4: Correlation Heatmap
plt.figure(figsize=(12,8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Heart Disease Dataset')
plt.tight_layout()
plt.savefig('outputs/plots/correlation_heatmap.png')
plt.show()
```

```python
# Cell 5: Distribution of Target Variable
plt.figure(figsize=(5,3))
sns.countplot(x='target', data=df, palette='Set2')
plt.title('Distribution of Target Variable')
plt.xlabel('Heart Disease Presence (1=Yes, 0=No)')
plt.ylabel('Count')
plt.tight_layout()
plt.savefig('outputs/plots/target_distribution.png')
plt.show()
```
